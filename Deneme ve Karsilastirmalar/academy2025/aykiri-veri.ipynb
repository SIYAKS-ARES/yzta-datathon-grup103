```python
import pandas as pd

train_df = pd.read_csv('train.csv')
```

```python
# Gerekli kütüphaneler
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot ile hızlı görselleştirme
plt.figure(figsize=(8, 4))
sns.boxplot(data=train_df, x='ürün fiyatı')
plt.title("Ürün Fiyatı - Boxplot (Aykırı Değer Kontrolü)")
plt.show()
```

```python
# IQR yöntemiyle alt ve üst sınırları belirle
q1 = train_df['ürün fiyatı'].quantile(0.25)
q3 = train_df['ürün fiyatı'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Aykırı değerleri filtrele
outliers = train_df[(train_df['ürün fiyatı'] < lower_bound) | (train_df['ürün fiyatı'] > upper_bound)]

# Kaç aykırı değer var?
print(f"Aykırı değer sayısı: {outliers.shape[0]}")
print(outliers[['ürün', 'ürün fiyatı']].sort_values(by='ürün fiyatı', ascending=False).head())
```
